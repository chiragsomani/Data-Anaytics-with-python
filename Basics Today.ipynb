```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pokemon.csv')
```

```python
df.head()
```

```python
##slicing columns
## colon -> everything
subset = df.loc[:, ['against_dark', 'against_fire']]
```

```python
subset.head(3)
```

```python
## using iloc command
## 2, 7 are column number, here integers can also be given e.g. -1 to select last column
subset = df.iloc[:, [2, 7]]
```

```python
subset.head(3)
```

```python
subset = df.iloc[:, list(range(2))]
subset.head(3)
```

```python
## accessing a particular cell
## to get abilities of zeroth row in 2 different ways 
print(df.loc[0, 'abilities'])
print(df.iloc[0, 0])
```

```python
## accessing particular cells / subsetting multiple rows and columns
df.iloc[[0, 99], list(range(7))]
```

```python
## MORE PREFERRED
df.loc[[0, 99], ['abilities', 'against_bug', 'against_dragon', 'against_electric', 'against_fairy']]
```

```python
df.loc[5:14, ['abilities', 'against_bug', 'against_dragon', 'against_electric', 'against_fairy']]
```

```python
## Grouped MEAN
```

```python
df.groupby('type1')['speed'].mean()
```

```python
## GROUPED MEAN MULTI Group
```

```python
df.groupby(['type1', 'type2'])['speed', 'weight_kg'].mean()
```

```python
##Flatten Dataframe
df.groupby(['type1', 'type2'])['speed', 'weight_kg'].mean().reset_index().head(15)
```

```python
## nunique -> to get unique count on a Pandas Series
df.groupby('generation')['type1'].nunique()
```
